```python
import sqlite3
from datetime import datetime

# 数据库路径
DB_PATH = "data/ehrsql/eicu.db"

# SQL查询
query = """
SELECT patient.hospitaladmitsource 
FROM patient 
WHERE patient.uniquepid = '035-166' 
  AND datetime(patient.hospitaladmittime, 'start of year') = datetime(current_time, 'start of year', '-1 year') 
ORDER BY patient.hospitaladmittime DESC 
LIMIT 1
"""

def execute_query(db_path, query):
    """
    执行SQL查询并返回结果
    
    Args:
        db_path: 数据库文件路径
        query: SQL查询语句
        
    Returns:
        查询结果（列表）
    """
    try:
        # 连接数据库
        conn = sqlite3.connect(db_path)
        cursor = conn.cursor()
        
        print("🔍 Executing query:")
        print(query)
        print("\n" + "=" * 60)
        
        # 执行查询
        cursor.execute(query)
        
        # 获取结果
        results = cursor.fetchall()
        
        # 打印结果
        if results:
            print(f"✅ Query returned {len(results)} row(s):")
            for i, row in enumerate(results, 1):
                print(f"   Row {i}: {row}")
                
            # 如果只有一个结果，直接返回值
            if len(results) == 1 and len(results[0]) == 1:
                answer = results[0][0]
                print(f"\n📋 Answer: {answer}")
                return answer
        else:
            print("⚠️  Query returned no results")
            return None
        
        # 关闭连接
        conn.close()
        
        return results
        
    except sqlite3.Error as e:
        print(f"❌ Database error: {e}")
        return None
    except Exception as e:
        print(f"❌ Error: {e}")
        return None



result = execute_query(DB_PATH, query)
print(f"\n🎯 Final result: {result}")
```
